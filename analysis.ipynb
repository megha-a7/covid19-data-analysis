```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print("Modules imported!!!")
```

```python
#importing COVID19 dataset
corona_dataset = pd.read_csv("covid19_Confirmed_dataset.csv")
corona_dataset.head(10)
```

```python
#check the shape of the dataset
corona_dataset.shape
```

```python
corona_dataset.drop(["Lat", "Long"], axis=1, inplace = True)
```

```python
corona_dataset.head(10)
```

```python
#aggregrate the rows by columns - to have data from each country
corona_dataset_aggregated = corona_dataset.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
#visualization
corona_dataset_aggregated.loc["China"]
```

```python
#calculating a good measure - finding a good measure represented as a number, describing the spread of the virus in a country
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["Italy"].plot()
plt.legend()
```

```python
#first three days
corona_dataset_aggregated.loc["China"][:3].plot()
```

```python
#calculating and plotting the first derivative of the curve - change in infection rate day after day in the period
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
#calculating the maximum infection rate for China, Italy and Spain
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
corona_dataset_aggregated.loc["Italy"].diff().max()
```

```python
corona_dataset_aggregated.loc["Spain"].diff().max()
```

```python
#calculating maximum infection rate for all countries
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries :
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["max_infection_rate"]= max_infection_rates
```

```python
corona_dataset_aggregated.head()
```

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rate"])
```

```python
corona_data.head()
```

```python
#worldwide_happiness
happiness_report = pd.read_csv("worldwide_happiness_report.csv")
happiness_report.head()
```

```python
useless_col= ["Overall rank","Score","Generosity","Perceptions of corruption"]
```

```python
happiness_report.drop(useless_col, axis=1, inplace = True)
```

```python
happiness_report.head()
```

```python
#changing the indices of the dataframe
happiness_report.set_index("Country or region", inplace= True)
```

```python
happiness_report.head()
```

```python
corona_data.shape
```

```python
happiness_report.shape
```

```python
data = corona_data.join(happiness_report, how ="inner")
```

```python
data.head()
```

```python
#correlation matrics
data.corr()
```

```python
#plotting gdp vs maximum infection rate
x= data["GDP per capita"]
y = data["max_infection_rate"]
sns.scatterplot(x, np.log(y))
```

```python
#fitting a curve
sns.regplot(x, np.log(y))
```
